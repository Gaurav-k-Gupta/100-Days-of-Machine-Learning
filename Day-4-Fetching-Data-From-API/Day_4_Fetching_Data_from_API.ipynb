Import requests and pandas, then call the api using request.get() method

```python
import requests
import pandas as pd

url = "https://api.themoviedb.org/3/movie/top_rated?api_key=f50d62d1ab0104664863698788b258cc&page=2"

response = requests.get(url)

data = response.json()['results']
```

Convert the response into data frame

```python
pd.DataFrame(data)[['id','title','original_title','overview','popularity','release_date','adult','vote_average']]
```

Using loop to fetch data from multiple pages of the url

```python
df = pd.DataFrame()
```

```python
for i in range (1,100):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=f50d62d1ab0104664863698788b258cc&page={}'.format(i))
  temp = pd.DataFrame(response.json()['results'])[['id','title','original_title','overview','popularity','release_date','adult','vote_average']]
  df = pd.concat([df , temp],ignore_index=True)
```
