<h1>Logistic Regression</h1>

Logistic Regression classify data into two categories by drawing an optimized line.


```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=100 , n_features=2 , n_informative=1 , n_redundant=0 , n_classes=2 , n_clusters_per_class=1 , random_state=41 , hypercube=False , class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
Y
```

```python
plt.scatter(X[:,0] , X[:,1] , c=Y , cmap='winter' , s=100)
```

<h1>Perceptron Trick</h1>

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X,Y):
    X = np.insert(X,0,1,axis=1)

    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(500):
        j = np.random.randint(0 , 100)
        Y_hat = step(np.dot(X[j] , weights))
        weights = weights + lr*(Y[j] - Y_hat)*X[j]

    return weights[0] , weights[1:]
```

```python
intercept , coef = perceptron(X,Y)
```

```python
intercept , coef 
```

```python
m = -(coef[0] / coef[1])
b = -(intercept / coef[1])
```

```python
X_input = np.linspace(-3 , 3 , 100)
Y_input = m*X_input + b 
```

```python
plt.plot(X_input , Y_input , color='red' , linewidth=3)
plt.scatter(X[:,0] , X[:,1] , c=Y , cmap='winter' , s=100)
plt.ylim(-3,2)
```

```python
# Percepton trick is not very efficient way of applying logistic regression
# after classifying all the data, percepton trick does not care about the line and symmetricity.
# Sometimes, it becomes the case of overfitting
```
