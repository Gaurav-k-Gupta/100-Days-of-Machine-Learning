<h1> We are fetching books data from the website books.toscrape.com</h1>

<h2>Fetching the name of books</h2>

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

base_url = "https://books.toscrape.com/catalogue/page-{}.html"
name = []



for i in range(1,51):
  url = base_url.format(i)
  temp = requests.get(url,headers=headers).text
  soup = BeautifulSoup(temp,'lxml')
  h3_list = soup.find_all('h3')
  for i in h3_list:
    name.append(i.find('a')['title'])

name
```

<h2>Fetching rating of all the books</h2>

```python
count = 0
rating = []
for i in range(1,51):
  response = requests.get("https://books.toscrape.com/catalogue/page-{}.html".format(i),headers=headers).text
  html = BeautifulSoup(response,'lxml')
  rating_list = html.find_all('p')
  for j in rating_list:
    if(j['class'][0]=='star-rating'):
      rating.append(j['class'][1])
      count += 1

rating
# print(count)
# rating_list[0]['class'][0]
# print(html.prettify())
```

<h2>Fetching the price of books<h2>

```python
count = 0;
prices = []
for i in range(1,51):
  response = requests.get("https://books.toscrape.com/catalogue/page-{}.html".format(i),headers=headers).text
  html = BeautifulSoup(response,'lxml')
  p_list = html.find_all('p',class_='price_color')
  for item in p_list:
    prices.append(item.text)
    count += 1

prices
# print(count)
```

```python
df = pd.DataFrame({
    'Name' : name,
    'Rating' : rating,
    'Price' : prices
})

df
```
