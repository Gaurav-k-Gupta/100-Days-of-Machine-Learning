<h1>Opening a csv file from a URL</h1>


```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/Gaurav-k-Gupta/100-Days-of-Machine-Learning/refs/heads/main/Day-2-Working-with-CSV-files/california_housing_test.csv"
headers = {"User-Agent" : "Mozilla/5.0 (Macintosh;; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

<h1>Sep parameter</h1>

```python
# sep parameter is used to specify separation of the columns in your data
# names parameter is used to specify the column names
pd.read_csv('sample_data/california_housing_test.csv',sep=',',names=['col1','col2','col3','col4','col5','col6','col7','col8','col9'])
```

<h1>Index Col parameter</h1>

```python
# It removes the default indexing and the column you specify becomes the new index!
pd.read_csv('sample_data/california_housing_test.csv',index_col='longitude')
```

<h1>Header Parameter</h1>

```python
# The specified row in this parameter becomes the first row , that is column names
pd.read_csv('sample_data/california_housing_test.csv',header=1)
```

<h1>Use cols Parameter</h1>

```python
# Only those columns are shown which are specified in the usecols parameter
pd.read_csv('sample_data/california_housing_test.csv',usecols=['longitude','latitude','total_rooms'])
```

<h1>Skip rows / n rows parameter</h1>

```python
# skiprows skips the rows mentioned (Not the range )
# pd.read_csv('sample_data/california_housing_test.csv',skiprows=[1,4])
# nrows only shows the given number of rows
pd.read_csv('sample_data/california_housing_test.csv',nrows=100)
```

<h1> Skip Bad Lines</h1>

```python
# there might be some rows which has more or less values than number of columns, so it is used to skip those bad lines
pd.read_csv('sample_data/california_housing_test.csv',on_bad_lines='skip')
```

<h1>Dtype Paramaters</h1>

```python
# Used to change the data type of the specified column
pd.read_csv('sample_data/california_housing_test.csv',dtype={'total_rooms':int}).info()
```

<h1>na_values parameter</h1>

```python
# the passed value in the parameter becomes NaN
pd.read_csv('sample_data/california_housing_test.csv',na_values=['27.0'])
```

<h1>Loading a huge dataset in chunks</h1>

```python
# The data is divided into chunks of given size
dfs = pd.read_csv('sample_data/california_housing_test.csv',chunksize=1000)
```
