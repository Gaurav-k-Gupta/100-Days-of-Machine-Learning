<h1>Lasso Regression or L1 Regression</h1>

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=100 , n_features=1 , n_informative=1 , n_targets=1 , noise=20 , random_state=13)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2 , random_state=12)
```

```python
plt.scatter(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_pred = lr.predict(X_test)

from sklearn.metrics import r2_score
r2_score(Y_test,Y_pred)
```

```python
lr.coef_ , lr.intercept_
```

```python
plt.scatter(X,Y)
plt.plot(X,lr.predict(X) , color='red')
```

```python
from sklearn.linear_model import Lasso

alphas = [0 , 1 , 2 , 5 , 10 , 30]
plt.scatter(X,Y)

for i in alphas:
    ls = Lasso(alpha=i)
    ls.fit(X_train,Y_train)
    plt.plot(X,ls.predict(X), label='alpha={}'.format(i))
    Y_pred = ls.predict(X_test)
    print('r2 score for alpha = {} : '.format(i) , r2_score(Y_test,Y_pred))

plt.legend()
plt.show()
```

```python
# Lasso Regression key points
# The coeffiecients become zero on increasing the value of alpha
# Higher coefficients are affected more
# on increasing alpha => bias increases and variance decreases
```
