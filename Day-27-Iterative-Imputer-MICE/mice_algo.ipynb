<h1>Iterative Imputer or MICE Algorithm</h1>

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression,LinearRegression
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
df = df.iloc[:,0:-1]
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[4,2] = np.nan
```

```python
df
```

```python
df0 = df.copy()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())

# df0

df1 = df0.copy()

df1.iloc[1,0] = np.nan

X = df1.iloc[[0,2,3,4],1:3]
y = df1.iloc[[0,2,3,4],0]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df1.iloc[1,1:].values.reshape(1,2))

df1.iloc[1,0] = 23.14

df1.iloc[3,1] = np.nan
X = df1.iloc[[0,1,2,4],[0,2]]
y = df1.iloc[[0,1,2,4],1]
lr = LinearRegression()
lr.fit(X,y)
lr.predict(df1.iloc[3,[0,2]].values.reshape(1,2))

df1.iloc[3,1] = 11.06
df1.iloc[4,-1] = np.nan
X = df1.iloc[0:4,0:2]
y = df1.iloc[0:4,-1]
lr = LinearRegression()
lr.fit(X,y)
val = lr.predict(df1.iloc[4,0:2].values.reshape(1,2))
df1.iloc[4,-1] = val
df1 - df0
# df1
```
